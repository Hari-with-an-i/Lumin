```python
import unsloth
from unsloth import FastLanguageModel
import torch
max_seq_length = 4096  
dtype = torch.bfloat16 if torch.cuda.is_bf16_supported() else torch.float16  
load_in_4bit = True 
```

```python
import gradio as gr
import time
import torch
import nltk
from nltk import sent_tokenize
from sentence_transformers import SentenceTransformer, util

nltk.download('punkt')

print("Using device:", "cuda" if torch.cuda.is_available() else "cpu")

# Clear cache
torch.cuda.empty_cache()
```

```python
embedder = SentenceTransformer('all-MiniLM-L12-v2')
```

```python
def get_prompt_templates():
    """Return prompt templates for different strategies."""
    templates = {
        "cot": """
You are a legal assistant drafting a clause compliant with the Companies Act 2013 of India. Follow these steps:

1. Analyze: Understand the requirement: {user_input}.
2. Extract: Identify key phrases in the requirement (e.g., 'all kinds of discrimination and harassment') to include verbatim or closely paraphrase.
3. Identify: Recall relevant sections of the Companies Act 2013.
4. Draft: Write a clear, professional clause in formal legal language, ensuring key phrases from the requirement are included.
5. Validate: Ensure the clause aligns semantically with the requirement and complies with applicable laws.

Instructions:
- Use formal legal language.
- Include key phrases from the requirement to ensure close alignment.
- Reference specific sections where applicable.
- Do not include placeholder text (e.g., Your Name).
- Format as:
Clause Title: [Short title]
Clause: [Clause text]

Drafted Clause:
""",
        "few_shot": """
You are a legal assistant drafting clauses compliant with the Companies Act 2013. Use these examples to guide your response, ensuring the clause closely aligns with the user's requirement by including key phrases.

Example 1:
Requirement: Draft a clause for immediate termination due to gross misconduct.
Clause Title: TERMINATION FOR GROSS MISCONDUCT
Clause: 
The Company may terminate the employment of any employee immediately, without notice, for gross misconduct, including but not limited to theft, fraud, or willful violation of company policies, in accordance with the Companies Act 2013.

Example 2:
Requirement: Draft a clause for termination due to non-performance.
Clause Title: TERMINATION FOR NON-PERFORMANCE
Clause: 
The Employee may be terminated with immediate effect for consistent failure to meet performance standards, as outlined in the employment agreement, subject to compliance with applicable laws under the Companies Act 2013.

Requirement: {user_input}

Instructions:
- Draft a clause like the examples, ensuring compliance.
- Include key phrases from the requirement (e.g., 'all kinds of discrimination and harassment') verbatim or closely paraphrased.
- Use formal legal language.
- Do not include placeholder text.
- Format as:
Clause Title: [Short title]
Clause: [Clause text]

Drafted Clause:
""",
        "role_based": """
You are an expert legal advisor specializing in the Companies Act 2013 of India. Draft a clause for: {user_input}

Instructions:
- Use your expertise to ensure compliance with the Companies Act 2013.
- Incorporate key phrases from the requirement (e.g., 'all kinds of discrimination and harassment') verbatim or closely paraphrased to maximize alignment.
- Use formal legal language and reference relevant sections.
- Do not include placeholder text.
- Format as:
Clause Title: [Short title]
Clause: [Clause text]

Drafted Clause:
"""
    }
    return templates
```

```python
def evaluate_cosine_similarity(user_input, generated_clause):
    """Calculate cosine similarity between user input and generated clause."""
    embeddings = embedder.encode([user_input, generated_clause], convert_to_tensor=True)
    similarity = util.cos_sim(embeddings[0], embeddings[1]).item()
    return similarity
```

```python
def load_model():
    model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    )
    return model, tokenizer

def generate_clause(user_input, strategy, model=None, tokenizer=None):
    """Generate a clause using the specified strategy, ensuring complete sentences."""
    templates = get_prompt_templates()
    prompt_template = templates[strategy]

    prompt = prompt_template.format(user_input=user_input)

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=1000,  # Increased to avoid truncation
        do_sample=False,  # Deterministic output for closer alignment
        temperature=0.5,  # Lower for less creativity
        top_p=0.9,
        pad_token_id=tokenizer.eos_token_id
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract clause
    clause_start = response.find("Drafted Clause:")
    if clause_start != -1:
        clause_start += len("Drafted Clause:\n")
        clause = response[clause_start:].strip()
    else:
        clause = response.strip()

    # Post-process to ensure complete sentences
    sentences = sent_tokenize(clause)
    if sentences:
        last_sentence = sentences[-1]
        if not last_sentence.endswith(('.', '!', '?')):
            sentences = sentences[:-1]
        clause = ' '.join(sentences)

    return clause
```

```python
def draft_and_compare(user_input, reference_clause=None):
    """Draft clauses using multiple strategies, evaluate, and display results."""
    global model, tokenizer
    strategies = ["cot", "few_shot", "role_based"]
    results = []

    for strategy in strategies:
        clause = generate_clause(user_input, strategy, model, tokenizer)
        
        cosine_sim = evaluate_cosine_similarity(user_input, clause)
        
        results.append({
            "Strategy": strategy.replace("_", " ").title(),
            "Clause": clause,
            "Cosine Similarity": cosine_sim,
        })
    print("\nGenerated Clauses and Evaluations:")
    for result in results:
        print(f"\n{result['Strategy']} Strategy:")
        print(f"Clause:\n{result['Clause']}\n")
        print(f"Cosine Similarity: {result['Cosine Similarity']:.4f}")
```

```python
import gradio as gr
from transformers import GenerationConfig

def draft_and_compare(user_input, reference_clause=None):
    """Draft clauses using multiple strategies, evaluate, and display results."""
    global model, tokenizer
    strategies = ["cot", "few_shot", "role_based"]
    results = []

    for strategy in strategies:
        clause = generate_clause(user_input, strategy, model, tokenizer)
        
        cosine_sim = evaluate_cosine_similarity(user_input, clause)

        results.append({
            "Strategy": strategy.replace("_", " ").title(),
            "Clause": clause,
            "Cosine Similarity": cosine_sim,
        })
    outputs = ""
    
    for result in results:
        outputs += f"\nStrategy:{result['Strategy']}\n"
        outputs += f"\nClause:\n{result['Clause']}\n"
        outputs += f"\nCosine Similarity:\n{result['Cosine Similarity']:.4f}\n"
        
    return outputs

gr.Interface(
    fn=draft_and_compare,
    inputs=[
        gr.Textbox(label="Clause"),
    ],
    outputs=gr.Textbox(label="Generated Clause", lines=5, max_lines=20, autoscroll=True),
    title="Custom Clause Generation"
).launch()
```
